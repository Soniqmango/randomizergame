```python
import numpy as np

totaldone = 0
attempts = 0
successfulnum = []
successfulattempts = []
totalattempts = 0

try:
    numberoftimes = int(input("How many numbers would you like to simulate? \n"))
except TypeError:
    print("Please only type an integer.")
    numberoftimes = int(input("How many numbers would you like to simulate? \n"))

try:
    difference = float(input("How big of a range do you want? \n"))
except TypeError:
    print("Please only type a number.")
    numberoftimes = int(input("How big of a range do you want? \n"))

realdiff = difference/2

try:
    while totaldone <= numberoftimes:  
        randomarr = np.random.rand(100)
        avg = np.mean(randomarr)
        attempts += 1
        if 0.5-realdiff < avg < 0.5+realdiff:
            successfulnum.append(avg)
            successfulattempts.append(attempts)
            totaldone += 1  
            attempts = 0  
except KeyboardInterrupt:
    print(f"Program cancelled. We managed to do {totaldone} numbers, and here's what we got: ")


for i in range (0,numberoftimes):
    x = successfulnum.pop(0)
    y = successfulattempts.pop(0)
    totalattempts += y
    print(f"The number {x} took {y} attempts to achieve.")
if numberoftimes>1:
    print(f"\nThis took {totalattempts} total attempts.")
    print(f"Each number took {totalattempts/numberoftimes} attempts to achieve on average.")
```

```python
import numpy as np

closestnumber = 0
attempts = 0 
goalnumber = float(input("What is your goal number (0-1)?\n"))

try:
    while True:
        randomarr = np.random.rand(10)
        avg = np.mean(randomarr)
        attempts += 1
        if abs(avg-goalnumber)<abs(closestnumber-goalnumber):
            closestnumber=avg
except KeyboardInterrupt:
    print(f"The closest number was {closestnumber}, it took {attempts} attempts.")
    print(f"The difference was {abs(closestnumber-goalnumber)}!")
```
